```python
image=cv2.imread(r'D:\DEEP LEARNING\open cv\invoice-sample.jpg')
h,w,c=image.shape
boxes=pytesseract.image_to_boxes(image)
for b in boxes.splitlines():
    b = b.split(' ')
    #I 128 983 163 1063 
    img = cv2.rectangle(image, 
                        (int(b[1]), h - int(b[2])), (int(b[3]), h - int(b[4])), 
                        (0, 255, 0), 
                        1)

cv2.imshow('img', img)
cv2.waitKey(0)
    
```
